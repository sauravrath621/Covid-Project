```python
import pandas as pd
covid=pd.read_csv('covid_19_clean_complete.csv', parse_dates=['Date'])
covid.head()
```

```python
covid['Active']=covid['Confirmed']-covid['Recovered']-covid['Deaths']
```

```python
covid
```

```python
top=covid[covid['Date']==covid['Date'].max()]
```

```python
top['Country/Region'].value_counts()
```

```python
top=top.groupby('Country/Region')['Confirmed','Deaths','Recovered','Active'].sum().sort_values(['Confirmed'],ascending=False).reset_index()
```

```python
top = top.groupby('Country/Region')['Confirmed','Deaths','Recovered','Active'].sum().sort_values(['Confirmed'],ascending=False).reset_index()
```

```python
top.head(20)
```

```python
top_20=top.head(20)
top_20
```
